```python
import pandas as pd
```

```python
import numpy as np
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies = movies.merge(credits, on="title")
```

```python
movies.head()
```

```python
movies = movies[
    [
        'movie_id',
        'title',
        'overview',
        'genres',
        'keywords',
        'cast',
        'crew'
    ]
]
```

```python
movies.dropna(inplace=True)
```

```python
movies.head()
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def convert3(obj):
    L = []
    count = 0

    for i in ast.literal_eval(obj):
        if count != 3:
            L.append(i['name'])
            count += 1
        else:
            break

    return L
```

```python
def fetch_director(obj):
    L = []

    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break

    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
for feature in ['genres', 'keywords', 'cast', 'crew']:
    movies[feature] = movies[feature].apply(
        lambda x: [i.replace(" ", "") for i in x]
    )
```

```python
movies['tags'] = (
    movies['overview']
    + movies['genres']
    + movies['keywords']
    + movies['cast']
    + movies['crew']
)
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(
    lambda x: " ".join(x)
)
```

```python
new_df['tags'] = new_df['tags'].apply(
    lambda x: x.lower()
)
```

```python
new_df.head()
```

```python
new_df['tags'][1]
```

```python
import pickle

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer(
    max_features=5000,
    stop_words='english'
)

vectors = cv.fit_transform(
    new_df['tags']
).toarray()

vectors.shape
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
def recommend(movie):

    movie_index = new_df[
        new_df['title'] == movie
    ].index[0]

    distances = similarity[movie_index]

    movies_list = sorted(
        list(enumerate(distances)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    print(f"\nMovies similar to {movie}:\n")

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```

```python
pickle.dump(
    new_df.to_dict(),
    open("movie_dict.pkl", "wb")
)
```

```python
pickle.dump(
    similarity,
    open("similarity.pkl", "wb")
)
```
